# Clasificarea imaginilor cu fete in functie de emotia sugerata - multi-label (angry, disgust, fear, happy, neutral, sad, surprise)

### Pasul 1 - incarcarea datelor si afisarea distributiei lor

```python
from utils import *

inputs, outputs = load_face_images()

plot_data_histograms(outputs, 'face images emotions')
```

### Pasul 2 - testarea datelor cu un clasificator deja antrenat

```python
from deepface import DeepFace

computed_outputs = []
for image in inputs:
    analysis = DeepFace.analyze(img_path=image, actions=('emotion',), enforce_detection=False)
    computed_output = []
    for emotion, value in analysis['emotion'].items():
        if value > 0.2:
            computed_output.append(emotion)
    computed_outputs.append(computed_output)
```

### Pasul 3 - Afisarea tuturor clasificarilor determinate de clasificator

```python
for output in computed_outputs:
    print(output)
```
